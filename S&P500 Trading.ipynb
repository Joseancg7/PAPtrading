```python
import numpy as np 
import pandas as pd 
import xlsxwriter  
import math 
```

```python
stocks = pd.read_excel('descarga.xlsx')
data=pd.DataFrame(stocks.iloc[2:])
data
```

```python
dataframe=data.iloc[0:,[0,1,2,3,4]]
dataframe
```

```python
portfolio_size=int(input('Size of your portfolio '))
```

```python
portfolio_size
```

```python
weight=data.iloc[0:,3]
```

```python
cantidad = weight.multiply((1/100)*portfolio_size)
dataframe['$ Invest']=cantidad
```

```python
dataframe['Number Shares Buy']=(dataframe['$ Invest']/dataframe['Price Index *']).round()
dataframe
```

```python
writer = pd.ExcelWriter('recommended_trades.xlsx', engine='xlsxwriter')
dataframe.to_excel(writer, sheet_name='Recommended Trades', index = False)
writer.save()
```
